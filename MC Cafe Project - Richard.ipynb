```python
# Initialize varaibles and constants including building the menu dictionary.
ans = "y"
order = {}
receipt = {}
menu = {"Coffee" : {"Espresso" : 1.5, "Americano" : 2.0, "Caffe latte" : 2.25, "Mocha" : 2.25, "Cappuchino" : 2.50, "Macchiato" : 2.50},
        "Tea" : {"White tea" : 1.5, "Green tea" : 2.0, "Oolong tea" : 2.25, "Black tea" : 2.25, "Earl grey" : 2.5, "Matcha" : 2.5},
       "Pastries" : {"Croissant" : 5.5, "Cinnamon roll" : 4.5, "Banana bread" : 4.25, "Pumpkin bread" : 3.75, "Sticky bun" : 4.25, "Apple turnover" : 3.75}}

# Define all required functions.

# Function to display the menu.
def display_menu() :
    for item in menu :
        print(item)
        for product in menu[item] :
            print(f"\t{product}\t",end="")
            if len(product) < 7 : print(f"\t",end="")
            print(f"${menu[item][product] :.2f}")
    print()

# Function to display a yes no question and return the validated result.
def get_yes_or_no(msg) :
    ans = "y"
    while ans == "y" :
        ans = input(msg)
        if ans[0].lower() != "n" and ans[0].lower() != "y":
            print("Valid answers are \"y\" or \"n\".")
            ans = "y"
        else :
            ans = ans[0].lower()
            return ans  

# Function to build an order by prompting the customer to enter an item and the quantity that they want.
    # Tell the customer when an item they attempted to order is not on the menu.
    # Ask the customer if they would like to order another item.
    # Once the customer is finished with the order input, send back the order as a new dictionary 
def build_order() :
    ans = "y"
    order = {}
    while ans == "y" :    
        item = input(f"Please enter the desired menu item:")    # Get the item name
        if (menu["Coffee"].get(item.capitalize()) == None and menu["Tea"].get(item.capitalize()) == None and
            menu["Pastries"].get(item.capitalize()) == None) :  # Check that the item name exists
            print("I'm sorry, {item} is not on the menu.")
        else :                                                  # Get the quantity of the item and validate it
            quantity = 0
            while quantity == 0 :
                quantity_str = input(f"Please enter the quantity of {item}s you would like to order:")
                if quantity_str.isdecimal() :
                    quantity = int(quantity_str)
                else :
                    print("Please enter only numbers.")
            if order.get(item.capitalize()) != None :                     #build the order dictionary
                order[item.capitalize()] += quantity
            else :
                order[item.capitalize()] = quantity                 
        ans = get_yes_or_no("Would you like to order another item(y/n)?")
    return order            

# Function to build a receipt dictionary from an order list.
def build_receipt(order) :
    receipt = {}
    for item in menu :
        for order_item in order :
            if menu[item].get(order_item) != None :
                receipt[order_item] = (order[order_item], order[order_item]*menu[item][order_item])
    return receipt

# Function to display a receipt from a receipt dictionary.
def display_receipt(rcpt) :
    subtotal = 0
    print(f"#\tItem\t\tCost")
    for item in rcpt :
        print(f"{rcpt[item][0]}\t{item}\t",end="")
        if len(item) < 7 : print(f"\t",end="")
        print(f"${rcpt[item][1] :.2f}")
        subtotal += rcpt[item][1]        #Create a running subtotal
    tax = subtotal * .06                 #Caculate tax and total
    total = subtotal * 1.06
    print("*"*31)  
    print(f"\tSubtotal\t${subtotal :.2f}")
    print(f"\tTax\t\t${tax :.2f}")
    print(f"\tTotal\t\t${total :.2f}")

# Main Program.

    # Welcome customers.

print (f"Welcome to MC Cafe!\n")

    # Ask the customer if they would like to see the menu. If yes, then display the menu items and prices.

ans = get_yes_or_no("Would you like to see the menu (y/n)?")
if ans == "y" : display_menu()

    # Ask the customer if they would like to order. 

ans = get_yes_or_no("Would you like to place an order(y/n)?")

    # If they want to order, build the order and display an order receipt that lists the items that they ordered and the quantity.
    # At the bottom of the receipt, display the subtotal, the amount in sales tax (tax is 6%), and the grand total (subtotal plus sales tax).

if ans == "y" :
    order = build_order()
    receipt = build_receipt(order)
    print()
    display_receipt(receipt)

# End the program and finish up by thanking the customer.

print (f"\nThank you for visiting MC Cafe!")
```
